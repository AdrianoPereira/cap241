```python
ano = [
    jan['desmatamento'].sum(),
    fev['desmatamento'].sum(),
    mar['desmatamento'].sum(),
    abr['desmatamento'].sum(),
    mai['desmatamento'].sum(),
    jun['desmatamento'].sum(),
    jul['desmatamento'].sum(),
    ago['desmatamento'].sum(),
    set_['desmatamento'].sum(),
    out['desmatamento'].sum(),
    nov['desmatamento'].sum()
]
```

```python
# Pie chart
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10]
# only "explode" the 2nd slice (i.e. 'Hogs')
explode = (0, 0.1, 0, 0)  
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```

```python
def getPie(month):
    aa = [int(a) for a in month.query('desmatamento > 60')['desmatamento']]
    bb = [a for a in month.query('desmatamento > 60')['cidade']]
    return aa, bb 
```

```python
fig1, ax1 = plt.subplots()
values, names = getPie(jun)

# explode = (0, 0.1, 0, 0)  

ax1.pie(values, labels=names, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  
plt.savefig('./desmatamento2017pie.png', format='png', dpi=600)
plt.tight_layout()
plt.show()
```
