```python
def convexHull(multipolygon):
    return cascaded_union([m for m in polygon]).convex_hull

def getPolygonCity(name):
    name = name.upper()
    query = "%s == '%s'"%(columns[0], name)
    return municipios_para.query(query)

def getPoints(city):
    try:
#         print(city)
        points = getPolygonCity(city)
        coords = None

        if type(list(points.geometry)[0]) == MultiPolygon:
            points = convexHull(list(points.geometry)[0])
#             print('Alter to ', type(points))
            coords = points.boundary.coords.xy
        else:
            coords = list(points['geometry'])[0].boundary.coords.xy

    #     coords = list(points['geometry'])[0].boundary.coords.xy
        line = ''
        for x in range(len(coords[0])):
            line += ' '+str(coords[1][x])+' '+str(coords[0][x])

        return line
    except IndexError:
        print('Error: ', city)

def writePoints(city):
    points = getPoints(city)
#     line = ''
    
#     for x,y in points:
#         line += ' '+str(x)
#         line += ' '+str(y)
    
    try:
        with open('./polygons/'+(city.lower())+'.txt', 'a') as f:
            f.write(points)
    except TypeError:
        print('Error to write ', city)
```

```python
city_names = list(municipios_para[columns[0]])
for city in city_names:
    try:
        city = city.lower()
        xx = list(getPolygonCity(city).geometry)[0].boundary.coords.xy[1]
        yy = list(getPolygonCity(city).geometry)[0].boundary.coords.xy[0]
        
        line = ''
        for i in range(len(xx)):
            line += ' {} {}'.format(xx[i], yy[i])
        
        with open('data_city/'+str(city)+'.txt', 'a') as f:
            f.write(line)
            
    except NotImplementedError:
        p = list(getPolygonCity(city).geometry)
        p = [pp for pp in p[0]]
        p = cascaded_union(p).convex_hull
        
        xx = p.boundary.coords.xy[1]
        yy = p.boundary.coords.xy[0]
                  
        line = ''
        for i in range(len(xx)):
            line += ' {} {}'.format(xx[i], yy[i])
        
        with open('data_city/'+str(city)+'.txt', 'a') as f:
            f.write(line)
        
    except SyntaxError:
        city.replace("'", "\'")
        municipios_para.query("NM_MUNICIP == \"{}\"".format(city.upper())))
        print(city)
```
